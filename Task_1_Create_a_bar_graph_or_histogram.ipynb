```python
for i in df1_copy.columns:
  plt.figure(figsize=(8,6))
  sns.countplot(x=i, hue='LUNG_CANCER', data=df1_copy, palette='viridis')
  plt.title(f'Frequency Distribution of {i} by LUNG_CANCER')
  plt.xticks(rotation=45)
  plt.show()
```

```python
column=['Gender', 'Air Pollution', 'Alcohol use',
       'Dust Allergy', 'OccuPational Hazards', 'Genetic Risk',
       'chronic Lung Disease', 'Balanced Diet', 'Obesity', 'Smoking',
       'Passive Smoker', 'Chest Pain', 'Coughing of Blood', 'Fatigue',
       'Weight Loss', 'Shortness of Breath', 'Wheezing',
       'Swallowing Difficulty', 'Clubbing of Finger Nails', 'Frequent Cold',
       'Dry Cough', 'Snoring']
```

```python
for i in column:
  plt.figure(figsize=(8,6))
  sns.countplot(x=i, hue='Level', data=df_copy, palette='Set1')
  plt.title(f'Frequency Distribution of {i} by Level')
  plt.xticks(rotation=45)
  plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df_copy, x='Age', hue='Level', bins=30, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
